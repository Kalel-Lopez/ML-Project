```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pandas_datareader.data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
start = pd.to_datetime('2016-1-1')
end = pd.to_datetime('2020-1-1')  
```

```python
df = pdr.get_data_yahoo('WMT', start, end)  
df
```

```python
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(df.index, df["Close"])
plt.title('Walmart Stock Price [2016 to 2020]')
plt.gcf().set_size_inches(12,8)
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.show()
```
